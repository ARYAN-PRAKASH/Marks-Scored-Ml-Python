## Making an Regresser who predicts the math score

```python
from sklearn.model_selection import ShuffleSplit
X=np.array([stud_df["reading"],stud_df["writing"]]).T
y=np.array(stud_df["math"])
sss=ShuffleSplit(n_splits=5,test_size=0.2,random_state=38)
```

```python
for train_index,test_index in sss.split(X,y):
    print("Train : ", train_index , "Test : " , test_index)
    X_train,X_test=X[train_index] , X[test_index]
    y_train,y_test=y[train_index] , y[test_index]
    
```

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer as Imputer
my_pipeline=Pipeline([
    ('imputer',Imputer(strategy="median")),
    ('std_scaler',StandardScaler())
])
```

```python

stud_t =  pd.DataFrame(np.array([stud_df["reading"],stud_df["writing"],stud_df["math"]) ).T
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```
